```python
import tensorflow as tf
import numpy
import matplotlib.pyplot as plt
import pandas as pd
```

# Evaluation

```python
# creating more data
x=tf.range(-100,100,4)
y=x+10
```

```python
plt.scatter(x,y)
```

# 3 sets
1. Training 70-80 percent model learns form this data
2. validation 10 to 15 percent model gets tuned on this data
3. test the mode lgets evaluated on this data 10 to 15 percent

```python
len(x)
```

```python
# training features
x_train=x[:40]
x_test=x[40:]
len(x_train),len(x_test)
```

```python
# training labels
y_train=y[:40]
y_test=y[40:]
len(y_train),len(y_test)
```

```python
# Reshape x_train and x_test to 2D
x_train = tf.expand_dims(x_train, axis=-1)  # Shape becomes (40, 1)
x_test = tf.expand_dims(x_test, axis=-1)    # Shape becomes (10, 1)
```

# visualize data

```python
plt.figure(figsize=(10,7))
plt.scatter(x_train,y_train,c="b",label="Training data")
plt.scatter(x_test,y_test,c="r",label="Testing data")
plt.legend()
plt.show()
```

```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, name="dense_input"),
    tf.keras.layers.Dense(100, activation="relu", name="dense_1"),
    tf.keras.layers.Dense(100, activation="relu", name="dense_2"),
    tf.keras.layers.Dense(1, name="dense_output")
])

model1.compile(loss="mae",
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               metrics=["mae"])

model1.fit(x_train, y_train, epochs=100, verbose=1)
```

# visualizig the model

```python
model1.summary()
```

1. Total Params: The total number of parameters (trainable + non-trainable) in the model.
2. Trainable Params: Parameters that the model learns during training (e.g., weights and biases).
3. Non-Trainable Params: Parameters that are fixed and not updated during training (e.g., frozen layers or pre-trained weights).
4. Optimized Params: Parameters being adjusted by the optimizer during training; equivalent to trainable parameters.

5. Weights: The parameters that define the strength of connections between neurons in a layer.
6. Bias: An additional parameter added to the weighted sum to shift the activation function's output.

```python
from tensorflow.keras.utils import plot_model
plot_model(model1,show_shapes=True)
```

Evaluation metrics
1. MAE (Mean Absolute Error): Measures the average magnitude of errors between predicted and actual values, treating all deviations equally regardless of direction.

2. MSE (Mean Squared Error): Measures the average of squared errors between predicted and actual values, giving more weight to larger deviations.

```python
loss1,mae1=model1.evaluate(x_test,y_test)
loss1,mae1
```

```python
predict1=model1.predict(x_test)
y_test,predict1
```

# Visualizeing the evaluation and prediction

```python
plt.figure(figsize=(10,7))
plt.scatter(x_test,y_test,c='b',label="real values")
plt.scatter(x_test,predict1,c="g",label="predicted values")
plt.title("Real values vs Predicted values")
plt.xlabel("x_test")
plt.ylabel("y_test")
plt.legend()
plt.show()
```

```python
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

model2.compile(loss="mae",
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.02),
               metrics=["mae"])

model2.fit(x_train, y_train,epochs=225, verbose=0)

```

# visualizeing the model

```python
model2.summary()
```

```python
plot_model(model2,show_shapes=True)
```

```python
loss2,mae2=model2.evaluate(x_test,y_test)
loss2,mae2
```

```python
predict2=model2.predict(x_test)
y_test,predict2
```

# Visualization of Model 2

```python
plt.figure(figsize=(10,7))
plt.scatter(x_test,y_test,c="b",label="Real Value")
plt.scatter(x_test,predict2,c='g',label="Predicted value")
plt.title("REAL VALUE VS PREDCTED VALUE OF MODEL 2")
plt.xlabel("x_test")
plt.ylabel("y_test")
plt.legend()
plt.show()
```

```python
model3=tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(1)
])

model3.compile(loss="mae",
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               metrics=["mae"])

model3.fit(x_train,y_train,epochs=250,verbose=0)
```

```python
model3.summary()
```

```python
plot_model(model3,show_shapes=True)
```

```python
mae3,loss3=model3.evaluate(x_test,y_test)
mae3,loss3
```

```python
predict3=model3.predict(x_test)
y_test,predict3
```

# visualize the model3

```python
plt.figure(figsize=(10,7))
plt.scatter(x_test,y_test,c="b",label="Real Values")
plt.scatter(x_test,predict3,c="g",label="Predicted Values")
plt.title("Real Values vs Predicted Values")
plt.xlabel("x_test")
plt.ylabel("y_test")
plt.legend()
plt.show()
```
